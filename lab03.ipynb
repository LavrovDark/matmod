**—Ä–µ—à–µ–Ω–∏–µ 1**

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
def dydt(y,t):
    k=0.6
    return y-y*y-k*y
t=np.linspace(0,10,100)
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten()
%matplotlib inline

plt.plot(t, solve, label = r"$y' = y-y*y-k*y$")
plt.grid(True)
plt.legend()
plt.show()
```

**—Ä–µ—à–µ–Ω–∏–µ 2 –ú–µ—Ç–æ–¥–æ–º –≠–π–ª–µ—Ä–∞**

```python
import numpy as np
from matplotlib import pyplot as plt

h=0.1
t2=np.arange(0,10+h,h)

def euler(f,y0,t,h):
        y=[]
        y.append(y0)
        for i in range(1,len(t)):
            yn=y[i-1]+h*f(y[i-1],t[i-1])
            y.append(yn)
        return y
    
euler=euler(dydt,1,t,h)

%matplotlib inline

plt.plot(t, euler, label = r"$euler$")
plt.plot(t, solve, label = r"$y' = y-y*y-k*y$")
plt.grid(True)
plt.legend()
plt.show()
```
