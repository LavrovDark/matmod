```python
from scipy.integrate import odeint
def dydt(y,t):
    k=0.01
    return k*y
```

```python
solve=odeint(dydt,y0=4000,t=years)
solve=solve.flatten()
%matplotlib inline

plt.plot(years, people, label = r"Население Берлина")
plt.plot(years, solve, label = r"Жесткая модель Мальтуса")
plt.grid(True)
plt.legend()
plt.show()
```

```python
def s(y):
    a=50000000
    b=350
    return a-b*y
def dydt2(y,t):
    return s(y)*y
```

```python
solve2=odeint(dydt2,y0=7000,t=years)
solve2=solve2.flatten()
%matplotlib inline

plt.plot(years, people, label = r"Население Берлина")
plt.plot(years, solve2, label = r"Мягкая модель Мальтуса")
plt.grid(True)
plt.legend()
plt.show()
```
